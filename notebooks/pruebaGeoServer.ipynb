```python
import requests
import geopandas as gpd
from io import BytesIO

url = (
    "https://proconservation-azzie-sirenically.ngrok-free.dev/geoserver/"
    "ne/ows?service=WFS&version=1.0.0&request=GetFeature"
    "&typeName=ne:GCPs&outputFormat=application/json"
)

user = "admin"          # cambia por el usuario real
password = "geoserver"  # cambia por la contrase침a real

r = requests.get(url, auth=(user, password))
gdf = gpd.read_file(BytesIO(r.content))
gdf.head()
```

```python
import folium

# Calcular el centro del mapa a partir del bounding box del GeoDataFrame
centroid = gdf.to_crs(epsg=4326).geometry.unary_union.centroid
center_lat, center_lon = centroid.y, centroid.x

# Crear el mapa base
m = folium.Map(location=[center_lat, center_lon], zoom_start=2)

# A침adir la capa del GeoDataFrame
folium.GeoJson(
    gdf.to_crs(epsg=4326),
    name="ne:GCPs",
    tooltip=folium.GeoJsonTooltip(fields=['id']), # Corrected: only include existing columns
).add_to(m)

folium.LayerControl().add_to(m)

m
```
