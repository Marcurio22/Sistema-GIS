```python
from pathlib import Path
import os

print("CWD =", os.getcwd())
print("多Existe data/raw?:", (Path("data/raw")).exists())
print("多Existe ../data/raw?:", (Path("../data/raw")).exists())
```

```python
import rasterio
red_path = "../data/raw/s2_B04.tif"
nir_path = "../data/raw/s2_B08.tif"

with rasterio.open(red_path) as r, rasterio.open(nir_path) as n:
    print("CRS iguales   :", r.crs == n.crs)
    print("Transform igual:", r.transform == n.transform)
    print("Shapes        :", r.width, r.height, "|", n.width, n.height)
```
